Function to get tracks by year. Running into API error where some songs in specific years are missing audio features

```python
def get_tracks_to_csv(year):
    artist_name = []
    artist_genre = []
    track_name = []
    track_id = []
    popularity = []
    
    for i in range(0,1000,50):
        track_results = sp.search(q=f'year:{year}', type='track', limit=50,offset=i)
        for i, t in enumerate(track_results['tracks']['items']):
            artist_name.append(t['artists'][0]['name'])
            artist_genre.append(sp.artist(track_results['tracks']['items'][i]['artists'][0]['uri'])['genres'])
            track_name.append(t['name'])
            track_id.append(t['id'])
            popularity.append(t['popularity'])
    track_df = pd.DataFrame({'artist_name' : artist_name, 'artist_genre' : artist_genre, 'track_name' : track_name, 'track_id' : track_id, 'popularity' : popularity})



    dict_list=[]
    for track in track_df['track_id']:
        dict_list.append(sp.audio_features(track)[0])
        
    #removes songs that have no audio features error
    dict_list_2 = []
    for i in range(len(dict_list)):
        if type(dict_list[i]) == dict:
            dict_list_2.append(dict_list[i]) 

    audio_feat_df = pd.DataFrame(dict_list_2)

    merged_df = track_df.merge(audio_feat_df, left_on='track_id', right_on='id')
    merged_df = merged_df.drop_duplicates(subset = ['track_id'], keep = 'last')
    merged_df.to_csv(f'./data/tracks_{year}.csv', index = False)
    return
```

```python
years = []
for i in range(2012, 2023):
    years.append(i)
    
for year in years:
    get_tracks_to_csv(year)
    print(f'{year} data collected')
```
